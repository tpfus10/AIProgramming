```python
import pandas as pd
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')

print(pop.head()); print(areas.head()); print(abbrevs.head())

#pop, abbrevs 조인
merged = pd.merge(pop, abbrevs, how='outer',
                  left_on='state/region', right_on='abbreviation')
merged = merged.drop('abbreviation', axis=1) #(삭제할 행/열의 이름, 행/열 지정:0은 행 1은 열)
merged.head()

#널값이 있는 행이 있는지 확인
merged.isnull().any() 

#해당 열에서 널값을 가진 행을 확인 
merged[merged['population'].isnull()].head()

#일치되는 값이 없는 지역을 확인
merged.loc[merged['state'].isnull(), 'state/region'].unique()

#abbreviation에서 누락된 데이터 삽입
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()

#merged, areas 조인
final = pd.merge(merged, areas, on='state', how='left')
final.head()

#널값이 있는 행이 있는지 확인
final.isnull().any()

#해당 열에서 널값을 가진 행을 확인
final['state'][final['area (sq. mi)'].isnull()].unique()

#널값을 삭제
final.dropna(inplace=True)
final.head()

#2010년 전체인구에 해당하는 데이터 선택
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head()

#인구 밀도를 계산하고 내림차순으로 정렬
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']
density.sort_values(ascending=False, inplace=True)
density.head() #상위 5개 항목
density.tail() #하위 5개 항목
```

```python
import pandas as pd
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')

print(pop.head()); print(areas.head()); print(abbrevs.head())

#pop, abbrevs 조인
merged = pd.merge(pop, abbrevs, how='outer',
                  left_on='state/region', right_on='abbreviation')
merged = merged.drop('abbreviation', axis=1) #(삭제할 행/열의 이름, 행/열 지정:0은 행 1은 열)
merged.head()

#널값이 있는 행이 있는지 확인
merged.isnull().any() 

#해당 열에서 널값을 가진 행을 확인 
merged[merged['population'].isnull()].head()

#일치되는 값이 없는 지역을 확인
merged.loc[merged['state'].isnull(), 'state/region'].unique()

#abbreviation에서 누락된 데이터 삽입
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()

#merged, areas 조인
final = pd.merge(merged, areas, on='state', how='left')
final.head()

#널값이 있는 행이 있는지 확인
final.isnull().any()

#해당 열에서 널값을 가진 행을 확인
final['state'][final['area (sq. mi)'].isnull()].unique()

#널값을 삭제
final.dropna(inplace=True)
final.head()

#2011년 전체인구에 해당하는 데이터 선택
data_total = final.query("year == 2011 & ages == 'total'")
data_total.head()

#2011년 18세 미만 인구에 해당하는 데이터 선택
data18 = final.query("year == 2011 & ages == 'under18'")
data18.head()

#각각 인구 밀도를 계산하고 내림차순으로 정렬
data_total.set_index('state', inplace=True)
density_total = data_total['population'] / data_total['area (sq. mi)']
density_total.sort_values(ascending=False, inplace=True)
density_total = density_total.to_frame().reset_index().head(5)

data18.set_index('state', inplace=True)
density_18 = data18['population'] / data18['area (sq. mi)']
density_18.sort_values(ascending=False, inplace=True)
density_18 = density_18.to_frame().reset_index().head(5)

#주별로 18세 이하와 주 전체 모두 인구밀도가 높은 상위 5개주를 2011년을 대상으로 검색
high_density = pd.merge(density_total, density_18, how='inner',
                        left_index=True, right_index=True)
high_density.head()
```

```python
import pandas as pd
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')

print(pop.head()); print(areas.head()); print(abbrevs.head())

#pop, abbrevs 조인
merged = pd.merge(pop, abbrevs, how='outer',
                  left_on='state/region', right_on='abbreviation')
merged = merged.drop('abbreviation', axis=1) #(삭제할 행/열의 이름, 행/열 지정:0은 행 1은 열)
merged.head()

#널값이 있는 행이 있는지 확인
merged.isnull().any() 

#해당 열에서 널값을 가진 행을 확인 
merged[merged['population'].isnull()].head()

#일치되는 값이 없는 지역을 확인
merged.loc[merged['state'].isnull(), 'state/region'].unique()

#abbreviation에서 누락된 데이터 삽입
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()

#merged, areas 조인
final = pd.merge(merged, areas, on='state', how='left')
final.head()

#널값이 있는 행이 있는지 확인
final.isnull().any()

#해당 열에서 널값을 가진 행을 확인
final['state'][final['area (sq. mi)'].isnull()].unique()

#널값을 삭제
final.dropna(inplace=True)
final.head()

#2010년 18세 미만 인구에 해당하는 데이터 선택
data18_2010 = final.query("year == 2010 & ages == 'under18'")
data18_2010.head()

#2011년 18세 미만 인구에 해당하는 데이터 선택
data18_2011 = final.query("year == 2011 & ages == 'under18'")
data18_2011.head()

#2012년 18세 미만 인구에 해당하는 데이터 선택
data18_2012= final.query("year == 2012 & ages == 'under18'")
data18_2012.head()

#각각 인구 밀도를 계산하고 내림차순으로 정렬
data18_2010.set_index('state', inplace=True)
density18_2010 = data18_2010['population'] / data18_2010['area (sq. mi)']
density18_2010.sort_values(ascending=False, inplace=True)
density18_2010 = density18_2010.to_frame().reset_index()

data18_2011.set_index('state', inplace=True)
density18_2011 = data18_2011['population'] / data18_2011['area (sq. mi)']
density18_2011.sort_values(ascending=False, inplace=True)
density18_2011 = density18_2011.to_frame().reset_index()

data18_2012.set_index('state', inplace=True)
density18_2012 = data18_2012['population'] / data18_2012['area (sq. mi)']
density18_2012.sort_values(ascending=False, inplace=True)
density18_2012 = density18_2012.to_frame().reset_index()

# 'population' 열을 기준으로 세 데이터프레임을 비교하는 조건 설정
condition = (data18_2010['population'] < data18_2011['population']) & (data18_2011['population'] < data18_2012['population'])

# 조건을 충족하는 행 선택
selected_regions = data18_2010[condition].drop_duplicates(subset=['state/region'])

# 결과 확인
print(selected_regions['state/region'])
```
