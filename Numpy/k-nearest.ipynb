```python
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D: projection='3d'        

# 3D  
rand = np.random.RandomState(42)
X = rand.rand(20, 3)

#  
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 3D  
#ax.scatter(X[:, 0], X[:, 1], X[:, 2], s=100)

#      
dist_sq_3d = np.sum((X[:, np.newaxis, :] - X[np.newaxis, :, :]) ** 2, axis=-1)

#   3 partition
k = 2
nearest_partition = np.argpartition(dist_sq_3d, k+1, axis=1)

#     
drawn_dots = []
drawn_lines = []

#     scatter
for j in nearest_partition[0, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='purple', s=100)
    ax.plot(*zip(X[j], X[0]), color='purple')
    drawn_dots.append((0, j))

for j in nearest_partition[2, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='orange', s=100)
    ax.plot(*zip(X[j], X[2]), color='orange')
    drawn_dots.append((0, j))

for j in nearest_partition[3, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='yellow', s=100)
    ax.plot(*zip(X[j], X[3]), color='yellow')
    drawn_dots.append((0, j))

for j in nearest_partition[4, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='blue', s=100)
    ax.plot(*zip(X[j], X[4]), color='blue')
    drawn_dots.append((0, j))

for j in nearest_partition[5, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='red', s=100)
    ax.plot(*zip(X[j], X[5]), color='red')
    drawn_dots.append((0, j))

for j in nearest_partition[6, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='green', s=100)
    ax.plot(*zip(X[j], X[6]), color='green')
    drawn_dots.append((0, j))

for j in nearest_partition[7, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='pink', s=100)
    ax.plot(*zip(X[j], X[7]), color='pink')
    drawn_dots.append((0, j))

for j in nearest_partition[8, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='black', s=100)
    ax.plot(*zip(X[j], X[8]), color='black')
    drawn_dots.append((0, j))

for j in nearest_partition[9, :k+1]:
    #    
    # scatter     
    ax.scatter(X[j, 0], X[j, 1], X[j, 2], color='grey', s=100)
    ax.plot(*zip(X[j], X[9]), color='grey')
    drawn_dots.append((0, j))

```
